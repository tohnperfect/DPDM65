### แทนที่ missing ด้วยค่า mean ของกลุ่ม

````python
set(unit_df['propertytype_name_en'])
````

#### ลองทำ กลุ่ม Apartment ก่อน

เลือก data ตามกลุ่ม

````python
unit_df[unit_df['propertytype_name_en']=='Apartment'].head()
````

หาค่า mean ของ price_min

````python
unit_df[unit_df['propertytype_name_en']=='Apartment']['price_min'].mean()
````

แทนค่า missing ของกลุ่ม

````python
unit_df_apart = unit_df[unit_df['propertytype_name_en']=='Apartment'].fillna({'price_min' : unit_df[unit_df['propertytype_name_en']=='Apartment']['price_min'].mean()})
````

````python
unit_df_apart.isnull().any()
````

#### ทำ Condo

````python
unit_df_condo = unit_df[unit_df['propertytype_name_en']=='Condo'].fillna({'price_min' : unit_df[unit_df['propertytype_name_en']=='Condo']['price_min'].mean()})
````

````python
unit_df_condo.head()
````

````python
unit_df_condo.describe()
````

````python
unit_df_apart.head()
````

## ต่อตารางในแนวแกน Y (เพิ่มจำนวนข้อมูล)

````python
fused_df = pd.concat([unit_df_apart,unit_df_condo])
````

````python
fused_df
````

#### วนลูปทำทุกกลุ่ม

````python
unit_df_temp0 = unit_df[unit_df['propertytype_name_en']==list(set(unit_df['propertytype_name_en']))[0]].fillna({'price_min' : unit_df[unit_df['propertytype_name_en']==list(set(unit_df['propertytype_name_en']))[0]]['price_min'].mean()})

for group in list(set(unit_df['propertytype_name_en']))[1:]:
    print(group)
    unit_df_temp1 = unit_df[unit_df['propertytype_name_en']==group].fillna({'price_min' : unit_df[unit_df['propertytype_name_en']==group]['price_min'].mean()})
    unit_df_temp0 = pd.concat([unit_df_temp0,unit_df_temp1])
````
