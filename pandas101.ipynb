# ราคาเฉลี่ยของโครงการบ้านที่ขอนแก่น

````python
project_df2['province_name_th']=='ขอนแก่น'
````

ใช้ & -> and และ | -> or ในการรวมค่าความจริงที่เป็น list

````python
(project_df2['province_name_th']=='ขอนแก่น') & (project_df2['propertytype_name_th']=='บ้าน')
````

````python
project_df2[(project_df2['province_name_th']=='ขอนแก่น') & (project_df2['propertytype_name_th']=='บ้าน')].describe()
````

Quiz แบบ เจาะจงคน

1.   โครงการที่ขอนแก่น กับ อุดร อันไหนแพงกว่ากัน
2.   เปรียบเทียบจำนวนบ้านในโครงการเฉลี่ยของโครงการบ้านที่ ขอนแก่น กับ กทม
3.   เปรียบเทียบราคาโครงการบ้าน+คอนโด ในจังหวัดขอนแก่น และ จังหวัดโคราช
4.   ~~(เก็บไว้ก่อน) จังหวัดมีข้อมูลโครงการบ้าน (คอนโด) มากที่สุด ~~
5.   ราคาเฉลี่ยของโครงการใน อุดร ขอนแก่น โคราช เทียบกับ เชียงใหม่ เชียงราย
6.   ให้สร้าง boxplot ของราคาบ้านในจังหวัดขอนแก่น เทียบกับ คอนโด
7.   บ้านในภาค เหนือ ใต้ อีสาน ที่ไหนแพงสุด ถูกสุด (เลือกจังหวัดที่เป็นตัวแทนภาคมา ภาคละ 2 จังหวัด)
8.   ความหลากหลายของโครงการในขอนแก่น เทียบกับ ชลบุรี (property type name)
